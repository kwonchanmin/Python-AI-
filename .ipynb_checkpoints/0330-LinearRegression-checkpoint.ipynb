```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

df = pd.DataFrame({
 '공부시간(x)' : [1,2,3,4,5,7,8,10,12,13,14,15,18,20,25,28,30],
 '시험점수(t)' : [5,7,20,31,40,44,46,49,60,62,70,80,85,91,92,97,98]
})
# 공부시간 = 독립변수 , 시험점수 = 종속변수
display(df.head())

# 그래프로 데이터의 분포를 알고 싶다.
# scatter()를 이용하면 된다.

# plt.scatter(독립변수 , 종속변수)
plt.scatter(df['공부시간(x)'], df['시험점수(t)'])
# 우리가 찾고 싶은건 classical linear regression model
# 현재 데이터에 대해서는 1차 직선으로 표현
# y = ax +b
# 직선을 그릴때는 plot()을 이용하면 그릴 수 있다.
# plt.plot(x데이터,t데이터)
# plt.plot(df['공부시간(x)'],2*df['공부시간(x)']+3, color='r')
# plt.plot(df['공부시간(x)'],5*df['공부시간(x)']-7, color='g')
# plt.plot(df['공부시간(x)'],1*df['공부시간(x)']+8, color='b')
plt.plot(df['공부시간(x)'],4*df['공부시간(x)']-10, color='m')

plt.show()
#머신러닝에서의 학습은 결국 계수를 찾아내는 과정이다.
```

```python
import numpy as np
# tensorflow library가 import 되어야 한다.
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Flatten, Dense
from tensorflow.keras.optimizers import SGD

#Training Data Set
x_data = np.array([1,2,3,4,5]).reshape(-1,1) # 5행 1열로 
t_data = np.array([3,5,7,9,11]).reshape(-1,1)

# Model
keras_model = Sequential() # 그림에서 model 사각형을 의미

# Layer 생성
input_layer = Flatten(input_shape=(1,))
output_layer = Dense(1,activation='linear')

# Model에 Layer를 추가
keras_model.add(input_layer)
keras_model.add(output_layer)

#Model 동작 설정
keras_model.compile(optimizer=SGD(learning_rate=1e-2), # 내가 알아내야할 값이다. 처음엔 잘 모르기 때문에 
                   loss='mse')

# Model 학습
keras_model.fit(x_data,
               t_data,
               epochs=2000) # epochs는 끝이 안나고 w,b가 갱신할 횟수
```

```python
result = keras_model.predict([[10]]) # 10에 대한 예측값을 알고 싶을 땐 10을 넣는 것이 아니라 
print(result)
```
