```python
# 필요한 module을 불러들여요!

import numpy as np
import pandas as pd
# 학습 데이터와 테스트 데이터를 분리해야 해요!(7:3 혹은 8:2 분리)
from sklearn.model_selection import train_test_split
# 정규화
from sklearn.preprocessing import MinMaxScaler
# 우리 데이터에는 결측치와 이상치가 없어요
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Flatten, Dense
from tensorflow.keras.optimizers import Adam
# 다중분류를 하다보니... one-hot encoding 작업을 해야 해요!
import tensorflow as tf
```

```python
# Raw Data Loading

df = pd.read_csv('./data/bmi.csv', skiprows=3)
display(df.head(), df.shape) # (20000, 3)

# 결측치와 이상치는 여기서는 따로 존재하지 않네, 다행히도
# 이상치 zscore, turkey defence



```

```python
# 데이터를 분리할 꺼에요!
# 학습데이터와 테스트데이터를 분리할꺼에요!

x_data_train, x_data_test, t_data_train, t_data_test = \
train_test_split(df[['height','weight']].values,
                df['label'].values,
                test_size=0.3)

print(x_data_train, x_data_train.shape) # (14000,2)

# 정규화 진행! 수치들 스케일조정하자
scaler = MinMaxScaler()
scaler.fit(x_data_train)

x_data_train_norm = scaler.transform(x_data_train)
x_data_test_norm = scaler.transform(x_data_test)
print(x_data_train_norm, x_data_train_norm.shape)

# One-Hot 처리를 해야 할 것 같아요!
print(t_data_train) # 현재 1차원인데 이거 one-hot으로 바꾸면 당연히 2차원으로 바뀌지!!!! 이 정도는 당연하다
# tf.one_hot(t_data_train, depth=)  # depth는 class의 개수 => 분류의 개수
print(tf.one_hot(t_data_train, depth=3).numpy())
t_data_train_onehot = tf.one_hot(t_data_train, depth=3).numpy()
t_data_test_onehot = tf.one_hot(t_data_test, depth=3).numpy()
```

```python
# Model
model = Sequential()

# Layer
model.add(Flatten(input_shape=(2,)))
model.add(Dense(3, activation='softmax'))

# Model 설정
model.compile(optimizer=Adam(learning_rate=1e-4),
             loss='categorical_crossentropy',
             metrics=['accuracy'])

# Model 학습
model.fit(x_data_train_norm,
         t_data_train_onehot,
         epochs=500,
         validation_split=0.2,
         verbose=1)
```

```python
# 이거는 애초에 학습 데이터가 같으니, 결과가 좋을 수 밖에 없지, 여기서는 val_accuract를 봐야됨
# 로스 낮아져야 좋은거는 당연하지
```

```python
# 다 끝났으면 모델 평가, evaulation 진행!
print(model.evaluate(x_data_test_norm, t_data_test_onehot))
# 최종 정확도
# 0.971666693687439
```

```python
# 예측을 해 보아요!
predict_data = np.array([[176,62]])
scaled_predict_data = scaler.transform(predict_data)
print(model.predict(scaled_predict_data))
```
