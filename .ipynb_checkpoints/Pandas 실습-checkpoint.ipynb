```python
# 데이터 처리(분석)을 하기 위해서 우리가 사용하는 실제적인 module
# 기본적인 자료구조는 2개가 존재

# 1. Series : 1차원 ndarray를 기반으로 만든 자료구조
# 2. DataFrame : Series를 세로로 이어 붙여 만든 2차원 자료구조
# DataFrame은 series로 구성되어 있다.
#Pandas를 설치해야지 사용할 수 있다.
```

```python
#Series부터 알아보자
import numpy as np
import pandas as pd
```

```python
s = pd.Series([-1,5,10,99], dtype="float64")   # 이렇게 하면 int로 잡히는데 보통 우리는 연산을 실수타입으로 하니깐
print(s)
# 1차원이다!! 여기는 표현이 세로로 하네 옆에 0,1,2,3 은 인덱스 번호
# -1,5,10,99 이런 게 nd.array 에 들어가 있는데 여기는 좀 더 그게 업그레이된 느낌
arr = s.values
print(arr)
print(type(arr))
print(s.index) # stop=4 4는 포함 안함
```

```python
s = pd.Series([1,5,-10,30],dtype=np.float64,index=['caa','b','a','k'])
print(s)
# 혼동하지 마세요
print(s[0:3]) # 슬라이싱은 원본 보존!!!
print(s[0])
print(s['caa':'k']) # 앞도 포함하고 뒤도 포함해요! 내가 만든 인덱스는 전부 포함하네!!!!
```

```python
print(s[1]) # 5.0 기본적인 숫자 인덱스
print(s['b']) # 5.0 내가 부여한 인덱스
```

```python
s = pd.Series([1,5,-10,30],dtype=np.float64,index=[0,2,100,6])
print(s)
print(s[1]) # 이렇게 내가 지정한 인덱스가 숫자면 기본 숫자인덱스 사용불가
# 이래서 내가 지정한 인덱스 숫자로 잘 사용 안함
print(s[100])
```

```python
s = pd.Series([1,5,-10,30],dtype=np.float64,index=['c','b','a','k'])
print(s)
# 혼동하지 마세요
print(s[0:3]) # 슬라이싱은 원본 보존!!!
print(s[0])
print(s['c':'k']) # 앞도 포함하고 뒤도 포함해요! 내가 만든 인덱스는 전부 포함하네!!!!
print(s[s%2==0])
print(s.shape)
print(s.sum())
```

```python
# 이번에는 Series를 조금 다르게 만들어 보자
# dictionary를 이용해서 Series를 생성

my_dict = {'서울': 2000, '부산': 3000, '인천': 500}
print(my_dict) # {'서울': 2000, '부산': 3000, '인천': 500}
print(type(my_dict)) # <class 'dict'>

s = pd.Series(my_dict) # 이렇게 시리즈를 직접 사용하는 경우는 많지 않다. 
print(s)
```

```python
#Series는 1차원 ndarray에 사용자 지정 index를 추가한 자료구조
# 일반적으로 Series를 직접 만들어 사용하는 경우는 많지 않다.
# 대부분 DataFrame을 사용한다.( 얘가 훨씬 편하고 기능도 많다. )

# DataFrame은 Excel 과 같다.
```

```python
import numpy as np
import pandas as pd

# 하나의 key에 하나의 value 
my_dict = {'name': '홍길동', 'year':2020 }
s = pd.Series(my_dict)
print(s)

# 하나의 key에 value를 여러 개 주는 방법
my_dict = {'names':['홍길동','신사임당','강감찬'], 'year':[2020,2021,2022] } # 이놈은 Series로 못만듬
df = pd.DataFrame(my_dict)
print(df) #   names  year
         #0   홍길동  2020
         #1  신사임당  2021
         #2   강감찬  2022

# DataFrame을 사용할 때는 print 말고 display를 쓰는게 이쁘게 나온다.
display(df) # 결과물은 실행시켜서 보자

# 데이터를 하나 더 줘보자
my_dict = {'names':['홍길동','신사임당','강감찬'], 'year':[2020,2021,2022],
          'point':[3.0, 4.0, 2.2]} 
df = pd.DataFrame(my_dict)

display(df)

# 중간에 데이터 개수가 안맞게 설정되면 에러가 나온다. 
# 넣어줄 값을 아직 설정하지 못한 상황의 error를 방지하고 싶으면 못정해준 값의 공간에 np.nan을 적어주면 된다.
my_dict = {'names':['홍길동','신사임당','강감찬'], 'year':[2020,2021,2022],
          'point':[3.0, 2.2, np.nan]} 
df = pd.DataFrame(my_dict)

display(df)
print(df.shape) # (3, 3)
print(df.size)  # 9
print(df.index) # RangeIndex(start=0, stop=3, step=1)
print(df.values) # [['홍길동' 2020 3.0]
                 #  ['신사임당' 2021 2.2]
                 #  ['강감찬' 2022 nan]]
print(df.columns) # Index(['names', 'year', 'point'], dtype='object')
```

```python
# DataFrame이 제공하는 여러가지 함수들에 대해서 하나하나 알아보고 있는 중이다.
# 통계는 크게 2가지로 나뉜다
# 기술 통계, 분석 통계
# 분석통계는 통계쪽에 대한 수학적인 개념이 들어오는 것
# 기술 통계는 우리가 할려는 것으로, 작업을 하는 기술이 아닌 종이에 글자를 적는 기술(discribe)이다.
# 기술 통계는 평균 , 합, 편차. 분산, 표준편차, 사분위, 공붕산, 상관관계 등이 이러한 관계에 있다는 것을 보여준다
# 수학적 계산을 하는 것은 아니다. 
# pandas가 함수를 제공!
# sum(), mean()

#python의 list를 만들껀데 2차원으로 만들꺼기 때문에 중첩 list의 형태로 만들꺼다.
data = [[2,np.nan],
        [7,-3],
        [np.nan,np.nan],
       [1,-2]]
# 이걸 가지고 DataFrame을 만들어 볼꺼다.
df = pd.DataFrame(data, columns=['one','two'], index=['a','b','c','d'])
# display(df)

# numpy의 ndarray가 아니라 pandas의 DataFrame하고 있다.
# 차이가 있다. ndarray의 2차원 list와 DataFrame은 다르다.

# print(df.sum() # df안의 값을 모두 더하는 것이 아니다. numpy와 다르다.
                # sum() 안의 인자가 없으면 axis=0으로 간주 
                # default를 axis=0으로 인식
# print(df.sum(axis=1))

# print(df.mean()) # 마찬가지로 axis를 주지 않으면 NaN을 있는 걸로 여길 껀지 없는 걸로 여길껀지에 따라 계산값이 달라진다
                 # 실행해보면 NaN을 무시하고 없는 걸로 판단하고 계산한다. 
# print(df.mean(axis=0, skipna=False)) # skipna = NaN을 무시하지 말고 값을 포함시켜 계산 하라고 한다. 
                                     # 이렇게 되면 결과값은 NaN으로 나오게 되어 좋지 않다.
# NaN을 안나오게 할려면 값을 채우거나 NaN을 없애야 한다.
# 간단하게 생각하면 지우면 간단하다. 하지만 실수가 발생할 경우 돌이킬 수 없게 된다. 
# 결국 우리는 어찌됬건 NaN을 처리해야된다. 처리 방법은 2가지가 있다.
# 첫번째 방법은 삭제다. => 이 방법이 사실 제일 좋은 방법이다. 하지만 만약 삭제 시 데이터가 충분하지 않으면 문제가 발생한다.
# 그러면 데이터가 충분하다고 판단되는 때가 언제인가?
# 경우에 따라 다르다. 일반적으로 10만건이 기준!

# 다른 방법으로
# NaN을 적절하게 수정해서 사용해야 된다.(interpolation - 보간)
# NaN을 다른 값들을 다 더해 평균으로 나타내는 경우도 있고 , 
# 자기랑 가까이 있는 놈이 값이 비슷한 경우가 많은 것을 보고 비슷하게 바꾸는 경우도 있다. 
# 여러가지 방안을 이용해서 값을 수정할거다. 
mean_one = df['one'].mean()
# mean_one #3.3333333333333335
mean_two = df['two'].mean()
# 함수를 이용해서 NaN을 찾아서 값을 대체
df['one'] = df['one'].fillna(value=mean_one) # fillna : NaN을 찾아서 값을 채우라는 의미

display(df)
```

```python
df1 = pd.DataFrame({'a': ['a0','a1','a2','a3'],
                   'b': [1,2,3,4],
                   'c' : ['c0','c1','c2','c3']
                   })
display(df1)

df2 = pd.DataFrame({'b': ['b0','b1','b2','b3'],
                   'c' : ['c0','c1','c2','c3'],
                   'd' : ['d0','d1','d2','d3'],
                   'e' : ['e0','e1','e2','e3']
                   })
display(df2)

#  1.위 아래 방향으로 이어 붙이자
# 컬럼명을 기준으로 가져다가 붙인다.
# result1 = pd.concat([df1,df2], axis=0) # ([]) -> [] 안에는 이어붙일 이름들이 나온다. 뒤에는 axis 가 나온다.
# 위에 결과는 그냥 두 결과값을 고대로 이어 붙인걸로 나온다. 보기 좋지 않다.

# index가 의미 없기 때문에 ignore_index를 써서 깔끔하게 이어붙인다.  
result1 = pd.concat([df1,df2], axis=0, ignore_index=True)
display(result1) 
```

```python
df1 = pd.DataFrame({'a': ['a0','a1','a2','a3'],
                   'b': [1,2,3,4],
                   'c' : ['c0','c1','c2','c3']
                   })
display(df1)

df2 = pd.DataFrame({'b': ['b0','b1','b2','b3'],
                   'c' : ['c0','c1','c2','c3'],
                   'd' : ['d0','d1','d2','d3'],
                   'e' : ['e0','e1','e2','e3']
                   }, index=[2,3,4,5])
display(df2)

# 이번에 해볼 건 DataFrame을 가로로 붙일 거다 (열로 붙일 꺼다)
# index를 기준으로 가로 방향으로 결합합니다.
result = pd.concat([df1,df2], axis=1)
display(result)
```

```python
import numpy as np
import pandas as pd

data1 = { "학번" : [1,2,3,4],
          "이름" : ["홍길동","신사임당","강감찬","이순신"],
          "학년" : [2,4,1,3]}

data2 = { "학번" : [1,2,4,5],
          "학과" : ["CS","MATH","MATH","CS"],
          "학점" : [3.4,2.9,4.5,1.2]}

df1 = pd.DataFrame(data1)
df2 = pd.DataFrame(data2)

display(df1)
display(df2)

# inner join을 해보자
pd.merge(df1,df2, on="학번", how="inner") # merge : 결합한다는 의미, on은 기준을 의미, 
                                         #how는 어떤 조인을 할 것인지를 써주는 곳

# outer join 
pd.merge(df1,df2, on="학번", how="outer")

```

```python
# 컬럼명이 다를 경우 
import numpy as np
import pandas as pd

data1 = { "학번" : [1,2,3,4],
          "이름" : ["홍길동","신사임당","강감찬","이순신"],
          "학년" : [2,4,1,3]}

data2 = { "학생학번" : [1,2,4,5],
          "학과" : ["CS","MATH","MATH","CS"],
          "학점" : [3.4,2.9,4.5,1.2]}

df1 = pd.DataFrame(data1)
df2 = pd.DataFrame(data2)

# 컬럼명이 다를 경우 on 을 2번써서 실행 시켜 주면 된다.
display(pd.merge(df1,df2, left_on="학번", right_on="학생학번", how="inner"))
```
